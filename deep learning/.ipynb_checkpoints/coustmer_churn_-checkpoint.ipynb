```python
#ENCODING CATEGORICAL VARIABLE...
```

```python
for col in df1.columns:
  print(f'{col}: {df1[col].unique()}')
```

```python
#replacing no internet service and no phone servicwe by 'NO'...

df1.replace({'No internet service':'No','No phone service':'No'},inplace=True)
```

```python
df1.InternetService.replace('No','NO IS',inplace=True)
```

```python
df1.replace({'Yes':1,'No':0},inplace=True)
```

```python
df1.replace({'Female':1,'Male':0},inplace=True)
```

```python
#one hot encoding for categorical variables..
```

```python
df2 = pd.get_dummies(df1,columns=['InternetService','Contract','PaymentMethod',])
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
for col in df2.columns:
  print(f'{col}: {df2[col].unique()}')
```

```python
#we need to scale 3 columns 'monthly charges','tenure','total charges'.

col_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[col_to_scale] = scaler.fit_transform(df2[col_to_scale])
```

```python
df2.shape
```

```python
#splitting the datset for training and testing
```

```python
x=df2.drop('Churn',axis='columns')
y=df2['Churn']
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=10,train_size=0.8)
```

```python
def ANN_(x_train,x_test,y_train,y_test,los,weights):
    model = keras.Sequential([
        keras.layers.Dense(26,input_shape=(26,),activation='relu'),
        keras.layers.Dense(15,activation='relu'),
        keras.layers.Dense(1,activation='sigmoid')
    ])

    model.compile(optimizer='adam',
                  loss=los,
                  metrics=['accuracy'])
    if weights==-1:
      model.fit(x_train,y_train,epochs=100)
    else:
      model.fit(x_train,y_train,epochs=100,class_weight=weights)

    print('\nmodel score :',model.evaluate(x_test,y_test))
    y_preds = np.round(model.predict(x_test))
    print('\nclassification report\n',classification_report(y_test,y_preds))



```

```python
ANN_(x_train,x_test,y_train,y_test,los='binary_crossentropy',weights=-1)
```

```python
# 1.using under sampling....
```

```python
count_class0,count_class1 = df2.Churn.value_counts()

df_class0 = df2[df2['Churn']==0]
df_class1 = df2[df2['Churn']==1]
```

```python
count_class0,count_class1
```

```python
df_class_0_us = df_class0.sample(count_class1)
df_for_us = pd.concat([df_class_0_us,df_class1],axis=0)

print('after under sampling\n', df_for_us.Churn.value_counts())
```

```python
x_us = df_for_us.drop('Churn',axis='columns')
y_us = df_for_us['Churn']

x_train_us,x_test_us,y_train_us,y_test_us = train_test_split(x_us,y_us,random_state=2,stratify=y_us,test_size=0.20)
```

```python
y_train_us.value_counts()
```

```python
y_test_us.value_counts()
```

```python
ANN_(x_train_us,x_test_us,y_train_us,y_test_us,los='binary_crossentropy',weights=-1)
```

```python
#***************************************************************************************************************************
```

```python
# 2. over sampling....
```

```python
df_class_1_os = df_class1.sample(count_class0,replace=True)


df_for_os = pd.concat([df_class0,df_class_1_os],axis=0)

print('after under sampling\n', df_for_us.Churn.value_counts())
```

```python
x_os = df_for_os.drop('Churn',axis='columns')
y_os = df_for_os['Churn']

x_train_os,x_test_os,y_train_os,y_test_os = train_test_split(x_os,y_os,random_state=2,stratify=y_os,test_size=0.20)
```

```python
y_train_os.value_counts()
```

```python
y_test_os.value_counts()
```

```python

ANN_(x_train_os,x_test_os,y_train_os,y_test_os,los='binary_crossentropy',weights=-1)
```

```python
# 3. <<<<SMOTE>>>>
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
x_sm,y_sm = smote.fit_resample(x,y)
```

```python
y_sm.value_counts()
```

```python
x_train_sm,x_test_sm,y_train_sm,y_test_sm = train_test_split(x_sm,y_sm,random_state=2,stratify=y_sm,test_size=0.20)

ANN_(x_train_sm,x_test_sm,y_train_sm,y_test_sm,los='binary_crossentropy',weights=-1)
```
