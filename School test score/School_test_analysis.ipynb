```python
fig,axes = plt.subplots(5,1,figsize=(5,22))
i=0
for col in df[cat_var]:
    sn.countplot(df,x=col,order = df[col].value_counts().index,ax=axes[i])
    axes[2].tick_params(labelrotation=30,labelsize=6)
    i+=1
```

##### BIVARIATE ANALYSIS

```python
fig, axs = plt.subplots(5,3,figsize=(12,15))
i=0
for col in df.columns:
    df.groupby(col)['MathScore'].mean().sort_values(ascending=False).plot.bar(ax=axs[i,0],color='lightblue')
    #axs[i,0].set_title(col+" vs Mathscore")
    axs[i,0].tick_params(labelrotation=10)
    axs[2,0].tick_params(labelrotation=20,labelsize=7)
    i+=1
    if i==5:
        break

i=0       
for col in df.columns:
    df.groupby(col)['ReadingScore'].mean().sort_values(ascending=False).plot.bar(ax=axs[i,1],color='violet')
    #axs[i,0].set_title(col+" vs ReadingScore")
    axs[i,1].tick_params(labelrotation=10)
    axs[2,1].tick_params(labelrotation=20,labelsize=7)
    i+=1
    if i==5:
        break
    
    
i=0       
for col in df.columns:
    df.groupby(col)['WritingScore'].mean().sort_values(ascending=False).plot.bar(ax=axs[i,2],color='lightgreen')
    #axs[i,0].set_title(col+" vs WritingScore")
    axs[i,2].tick_params(labelrotation=10)
    axs[2,2].tick_params(labelrotation=20,labelsize=7)
    i+=1
    if i==5:
        break
    
```

LIGHT BLUE = VS MATHSCORE

VIOLET = VS READINGSCORE

LIGHT GREEN = VS WRITINGSCORE

1. Male students scored more in reading, Writing, Math compared to female students
2. In ethnic group category Group E had performed well in the tests
3. There is not much difference in the parentedu categorey
4. No big difference in LunchType
5. Test prepration completed students had scored better
